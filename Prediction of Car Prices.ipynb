# 4 Filter Data

The data frame shows 15 columns with continuous values and 10 columns with discrete values.

Discrete values provides limited options for analysis, with low sensitivity outcomes and little indication of sources of variation. Continuous values provides variety of analysis options that can offer high sensitivity outcomes and insight into the sources of variation.

Therefore, select only the columns with continuous values and import into a new data frame named **numeric_cars**.

```python
numeric_cars = cars.filter(['normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price'], axis=1)
numeric_cars.head(5)
```

# 5 Data Cleaning

There are missing values with the string "?" which needs to be replaced with "NaN in the **numeric_cars** data.

It means "not a number", a value that you get if you perform a calculation whose result can't be expressed as a number. Any calculations you perform with "NaN" will also result in "NaN". 

```python
import numpy as np
numeric_cars = numeric_cars.replace('?', np.NaN)
numeric_cars.head(5)
```

Upon checking the data type of the attributes in **numeric_cars**, we see that it has a mix of object, float and integer data types. 

It is advisable to convert these data types to a single data type as we might have to reference a data type in a variable of another type.

```python
numeric_cars.dtypes
```

Convert the data types of attributes in **numeric_cars** to "float".

```python
numeric_cars = numeric_cars.apply(pd.to_numeric, downcast = 'float', errors = 'coerce')
numeric_cars.dtypes
```

Report the number of null values "NaN" for each attribute (column).

```python
len(numeric_cars) - numeric_cars.count()
```

Since 'price' is the column used for prediction and it has 4 rows with missing 'price' values, remove them entirely and again report the number of missing values "NaN" for each column.

```python
numeric_cars = numeric_cars[np.isfinite(numeric_cars['price'])]
len(numeric_cars) - numeric_cars.count()
```

Find column mean for each column.

```python
numeric_cars.mean()
```

Replace missing values "NaN" in other columns using column means.

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.head(5)
```

Ensure that there's no more missing values "NaN"

```python
len(numeric_cars) - numeric_cars.count()
```

We will be using the k-nearest neighbors algorithm to determine the prices of the car. The k-nearest neighbors algorithm uses the euclidean distance to determine the closest neighbor.

However in **numeric_cars**, each column has a different scaling. For example, if we take row 2 and row 3, the 'peak-rpm' has a difference of 500, while the difference in 'width' is 0.7. The algorithm in this case will give extra weight towards the difference in 'peak-rpm'.

Therefore, it is important to normalize the dataset into a unit vector with values ranging from 0 to 1.

Minimum - Maximum Normalization:

$z_i=(x_i-min(x))/(max(x)-min(x))$  
where $x=(x_1,x_2,\cdots,x_n)$ and $z_i$ is now your $i_{th}$ normalised data.

Normalise all columnns to range from 0 to 1 except for the target column 'price' as it will be used for prediction.

```python
norm_cars=((numeric_cars-numeric_cars.min())/(numeric_cars.max()-numeric_cars.min()))*1
norm_cars["price"]=numeric_cars["price"]
norm_cars.head(5)
```

# 6 Univariate Model [KNeighborsRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html) with Fixed k Value

Since we are predicting the price of cars, we should use **KNeighborsRegressor** rather than **KNeighborsClassifier**. 

Before we apply **KNeighborsRegressor**, use **train_test_split** function to randomise the order of rows and split the data into training set (80%) and test set (20%) with the random seed as 1.

Fit a KNN model using default k value **KNeighborsRegressor()**.

For each column except `price`, train a model, return RMSE value and add to the dictionary `rmse_results`. 

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

train_cols = norm_cars.columns.drop('price')
X = norm_cars[train_cols]
y = norm_cars['price']
rmse_results = {}

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

for i in X:
  adj = KNeighborsRegressor()
  adj.fit(X_train[[i]], y_train)
  forecasts = adj.predict(X_test[[i]])
  rmse_results[i] = np.sqrt(mean_squared_error(y_test, forecasts))
```

```python
rmse_results
```

# 7 Outcomes

Report the rmse_results in an ascending order.

```python
for s in sorted(rmse_results, key=rmse_results.get):
         print(s,'', rmse_results[s])
```

Good predictors will point to lower RMSE values and 'horsepower' seems to be the best predictor with the lowest RMSE value, closer to the actual price.

'compression-ratio' seems to be the worst predictor with highest RMSE value, far from the actual price.
